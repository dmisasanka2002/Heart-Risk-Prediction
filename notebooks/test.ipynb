```python
import numpy as np
import joblib
```

```python
data = np.load('../data/data.npy')
target = np.load('../data/target.npy')
model = joblib.load('../models/model.pkl')
```

```python
model
```

```python
target_pred = model.predict(data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(list(range(target.shape[0])), target, label='y_true', c='b')
plt.scatter(list(range(target_pred.shape[0])),target_pred, label='y_pred', c='g')
plt.legend()
```

```python
plt.scatter(target, target_pred, c='b')
plt.plot(target, target, label='100% Accuracy', c='r')
plt.legend()
plt.title('Y true vs Y pred')
plt.xlabel('Y True')
plt.ylabel('Y Predicted')
```
