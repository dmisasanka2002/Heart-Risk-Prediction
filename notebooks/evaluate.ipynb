```python
import joblib
import numpy as np
```

```python
model = joblib.load("../models/model.pkl")
# model = joblib.load("../models/model_pipe.pkl")

model
```

```python
x_test = np.load('../data/test_data.npy')
y_test = np.load('../data/test_target.npy')
x_test.shape
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(list(range(y_test.shape[0])), y_test, label='y_true', c='b')
plt.scatter(list(range(y_test.shape[0])),y_pred, label='y_pred', c='g')
plt.legend()
```

```python
plt.scatter(y_test, y_pred, c='b')
plt.plot(y_test, y_test, label='100% Accuracy', c='r')
plt.legend()
plt.title('Y true vs Y pred')
plt.xlabel('Y True')
plt.ylabel('Y Predicted')
```
